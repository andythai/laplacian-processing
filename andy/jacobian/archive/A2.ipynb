```python
import numpy as np
np.set_printoptions(linewidth=1000)

from scipy.linalg import null_space
import sympy
```

```python
# Load A2 and print stats
A2 = np.load('A2.npy')
print("Laplacian matrix A with no correspondences:")
print("\tShape:", A2.shape)
print("\tRank: ", np.linalg.matrix_rank(A2))
print("\tNull space:", null_space(A2))

# Run RREF on A2
A2_sympy = sympy.Matrix(A2)
display(A2_sympy)
A2_rref, pivots = A2_sympy.rref()
print("RREF:")
display(A2_rref)
```

Test adding a correspondence row

```python
A3 = A2.copy()
row_to_zero = 5
A3[row_to_zero] = np.zeros(A3.shape[1])  # Zero out the first row
A3[row_to_zero, row_to_zero] = 1
print("Matrix A with added correspondence:")
print("\tShape:", A3.shape)
print("\tRank: ", np.linalg.matrix_rank(A3))
print("\tNull space:", null_space(A3))

# Run RREF
A3_sympy = sympy.Matrix(A3)
display(A3_sympy)
A3_rref, pivots = A3_sympy.rref()
print("RREF:")
display(A3_rref)
```
